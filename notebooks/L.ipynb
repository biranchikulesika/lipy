# Dataset Structure

The training pipeline expects each dataset to follow a consistent structure.

Each dataset is stored as a ZIP archive in the Hugging Face Dataset repository. During training, the selected archive is downloaded and extracted into the local `dataset/` directory.

Example:

```text
complete_dataset.zip
│
└── complete_dataset/
    ├── CONS_KA_001.png
    ├── CONS_KA_002.png
    ├── CONS_KA_003.png
    ├── CONS_KHA_001.png
    ├── CONS_KHA_002.png
    ├── ...
```

---

## File Naming Convention

Every image filename must follow this format:

```text
<CATEGORY>_<CHARACTER>_<UNIQUE_ID>.png
```

Example:

```text
CONS_KA_0d3acbed-b712-4e0f-9f5f.png
```

Where:

| Part | Description |
|------|-------------|
| `CONS` | Character category |
| `KA` | Character identifier |
| `0d3acbed-b712-4e0f-9f5f` | Unique image identifier |

The notebook automatically extracts the class label using only the first two parts of the filename.

For example:

```text
CONS_KA_xxxxx.png
        │
        ▼
Class = CONS_KA
```

---

## Multiple Dataset Support

The project supports multiple datasets stored in the same Hugging Face repository.

Example:

```text

```

# Step 3: Download Dataset from Hugging Face

## Objective

Download the selected dataset archive from the Hugging Face Dataset repository and extract it into the local project directory.

The dataset to download is controlled by the `HF_DATASET_NAME` configuration variable.

---

## Output

- Downloaded dataset archive
- Extracted dataset directory
- `DATASET_DIR`

---

## Notes

The notebook first looks for an authentication token in:

1. Environment variable (`HF_TOKEN`)
2. Google Colab Secrets

If no token is found, the dataset is downloaded anonymously.

````python
# =============================================================================
# Download Dataset from Hugging Face
# =============================================================================

import os
import zipfile

try:
    from google.colab import userdata
except ImportError:
    userdata = None


def get_hf_token():
    """
    Return the Hugging Face access token.

    Priority:
        1. Environment variable
        2. Google Colab Secrets
        3. Anonymous access
    """

    token = os.getenv("HF_TOKEN")

    if token:
        return token

    if userdata is not None:
        try:
            token = userdata.get("HF_TOKEN")
            if token:
                return token
        except Exception:
            pass

    return None


HF_TOKEN = get_hf_token()

# Create local dataset directory
DATASET_ROOT.mkdir(parents=True, exist_ok=True)

ZIP_FILENAME = f"{HF_DATASET_NAME}.zip"

print("=" * 60)
print("Downloading Dataset")
print("=" * 60)
print(f"Repository : {HF_DATASET_REPO_ID}")
print(f"Dataset    : {ZIP_FILENAME}")

zip_path = hf_hub_download(
    repo_id=HF_DATASET_REPO_ID,
    repo_type="dataset",
    filename=ZIP_FILENAME,
    token=HF_TOKEN,
)

print(f"\nDownloaded archive:")
print(zip_path)

print("\nExtracting dataset...")

with zipfile.ZipFile(zip_path, "r") as archive:
    archive.extractall(DATASET_ROOT)

DATASET_DIR = DATASET_ROOT / HF_DATASET_NAME

print("\nDataset ready.")
print(f"Location : {DATASET_DIR.resolve()}")
````

# Step 4: Verify Dataset

## Objective

Verify that the selected dataset was downloaded and extracted successfully.

This step confirms that:

- The dataset directory exists.

## Output

# Step 5: Filter Valid Character Classes

## Objective

Group images by character label and remove classes that do not contain enough training samples.

Very small classes can negatively affect model performance and produce unstable predictions.

---

## Output

- `valid_classes`
- Filtered dataset statistics

---

## Notes

The minimum number of required images per class can be adjusted using the configuration value defined in this step.

````python
# =============================================================================
# Filter Valid Character Classes
# =============================================================================

from collections import Counter

# Count images belonging to each class
class_counts = Counter()

for image_path in image_files:

    parts = image_path.stem.split("_")

    # Expected filename format:
    # CONS_KA_xxxxxxxxx.png
    if len(parts) < 2:
        continue

    class_name = f"{parts[0]}_{parts[1]}"
    class_counts[class_name] += 1

# Keep only classes with enough samples
valid_classes = sorted(
    [
        class_name
        for class_name, count in class_counts.items()
        if count >= MIN_IMAGES_PER_CLASS
    ]
)

invalid_classes = sorted(
    [
        class_name
        for class_name, count in class_counts.items()
        if count < MIN_IMAGES_PER_CLASS
    ]
)

print("=" * 60)
print("Character Class Statistics")
print("=" * 60)

print(f"Total Images              : {len(image_files):,}")
print(f"Total Classes Found       : {len(class_counts)}")
print(f"Valid Classes             : {len(valid_classes)}")
print(f"Filtered Classes          : {len(invalid_classes)}")
print(f"Minimum Images per Class  : {MIN_IMAGES_PER_CLASS}")

print("\nValid Classes")

for class_name in valid_classes:
    print(f"  {class_name:<20} {class_counts[class_name]:>4}")

if invalid_classes:
    print("\nFiltered Classes")

    for class_name in invalid_classes:
        print(f"  {class_name:<20} {class_counts[class_name]:>4}")

print("\nClass filtering completed successfully.")
````

# Step 6: Generate Label Mapping

## Objective

Convert character labels into numeric class IDs required for neural network training.

Each unique character receives a unique integer identifier.

Example:

```text
CONS_KA   → 0
CONS_KHA  → 1
CONS_GA   → 2
```

---

## Output

- `label_map`
- `reverse_label_map`

---

## Notes

These mappings are exported later as `labels.json` and are used by the backend during inference.

````python
# =============================================================================
# Generate Label Mapping
# =============================================================================

# Create mappings between class names and numeric IDs
label_map = {
    class_name: idx
    for idx, class_name in enumerate(valid_classes)
}

reverse_label_map = {
    idx: class_name
    for class_name, idx in label_map.items()
}

NUM_CLASSES = len(label_map)

print("=" * 60)
print("Label Mapping")
print("=" * 60)

print(f"Total Classes : {NUM_CLASSES}\n")

print("Sample Label Mapping")
print("-" * 60)

for class_name in valid_classes[:10]:
    print(f"{label_map[class_name]:>2}  →  {class_name}")

if NUM_CLASSES > 10:
    print("...")

print("\nLabel mapping generated successfully.")
````

# Step 7: Load and Preprocess Images

## Objective

Load every image from the selected dataset and prepare it for model training.

During preprocessing, each image is:

- Loaded from disk
- Resized to **64 × 64**
- Converted from grayscale to RGB
- Normalized to the range **[0, 1]**
- Assigned its corresponding numeric class label

---

## Output

- `X` : Image tensor
- `y` : Integer labels

---

## Notes

Images belonging to filtered classes are skipped automatically.

Only PNG, JPG, and JPEG files are processed.

````python
# =============================================================================
# Load and Preprocess Images
# =============================================================================

images = []
labels = []

skipped_files = 0

for image_path in image_files:

    # Expected filename:
    # CONS_KA_xxxxxxxxx.png

    parts = image_path.stem.split("_")

    if len(parts) < 2:
        skipped_files += 1
        continue

    class_name = f"{parts[0]}_{parts[1]}"

    # Skip filtered classes
    if class_name not in label_map:
        continue

    # Read image
    image = cv2.imread(str(image_path), cv2.IMREAD_GRAYSCALE)

    if image is None:
        skipped_files += 1
        continue

    # Resize
    image = cv2.resize(image, IMAGE_SHAPE)

    # Convert grayscale → RGB
    image = cv2.cvtColor(image, cv2.COLOR_GRAY2RGB)

    # Normalize pixel values
    image = image.astype(np.float32) / 255.0

    images.append(image)
    labels.append(label_map[class_name])

# Convert to NumPy arrays
X = np.asarray(images, dtype=np.float32)
y = np.asarray(labels, dtype=np.int32)

print("=" * 60)
print("Image Preprocessing Summary")
print("=" * 60)

print(f"Images Loaded     : {len(X):,}")
print(f"Images Skipped    : {skipped_files}")
print(f"Image Shape       : {X.shape}")
print(f"Label Shape       : {y.shape}")
print(f"Image Data Type   : {X.dtype}")
print(f"Pixel Value Range : {X.min():.1f} → {X.max():.1f}")

print("\nImage preprocessing completed successfully.")
````

# Step 8: Encode Labels

## Objective

Convert integer class labels into one-hot encoded vectors required for multi-class classification.

Example:

```text
Class 3

↓

[0, 0, 0, 1, 0, 0, ...]
```

---

## Output

- One-hot encoded labels

---

## Notes

TensorFlow uses one-hot encoded labels together with the `categorical_crossentropy` loss function during training.

````python
# =============================================================================
# Encode Labels
# =============================================================================

# Convert integer labels to one-hot encoded vectors
y = to_categorical(y, num_classes=NUM_CLASSES)

print("=" * 60)
print("Label Encoding Summary")
print("=" * 60)

print(f"Number of Classes : {NUM_CLASSES}")
print(f"Encoded Labels    : {y.shape}")
print(f"Data Type         : {y.dtype}")

print("\nExample")

sample_index = np.random.randint(len(y))

print(f"Class ID       : {np.argmax(y[sample_index])}")
print(f"One-Hot Vector : {y[sample_index]}")

print("\nLabel encoding completed successfully.")
````

# Step 9: Split Training and Testing Data

## Objective

Split the dataset into independent training and testing subsets.

The training set is used to learn model parameters, while the testing set evaluates how well the model generalizes to unseen handwriting.

---

## Output

- `X_train`
- `X_test`
- `y_train`
- `y_test`

---

## Notes

A fixed random seed is used to make the split reproducible.

````python
# =============================================================================
# Split Training and Testing Data
# =============================================================================

# Split dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=TEST_SIZE,
    random_state=RANDOM_STATE,
    stratify=np.argmax(y, axis=1),
    shuffle=True
)

# Compute class weights to handle class imbalance
train_labels = np.argmax(y_train, axis=1)

class_weights = compute_class_weight(
    class_weight="balanced",
    classes=np.arange(NUM_CLASSES),
    y=train_labels
)

class_weight_dict = {
    class_id: weight
    for class_id, weight in enumerate(class_weights)
}

print("=" * 60)
print("Dataset Split Summary")
print("=" * 60)

print(f"Training Samples : {len(X_train):,}")
print(f"Testing Samples  : {len(X_test):,}")
print(f"Training Shape   : {X_train.shape}")
print(f"Testing Shape    : {X_test.shape}")
print(f"Class Weights    : {len(class_weight_dict)} classes")

print("\nDataset split completed successfully.")
````

# Step 10: Configure Data Augmentation

## Objective

Generate slightly modified versions of training images during each epoch.

This improves the model's ability to generalize by exposing it to different variations of the same handwritten character.

Examples include:

- Rotation
- Translation
- Zoom
- Shear
- Brightness variation

---

## Output

- Configured image generator (`datagen`)

---

## Notes

Data augmentation is applied only to the training dataset.

The testing dataset remains unchanged to ensure an unbiased evaluation.

````python
# =============================================================================
# Configure Data Augmentation
# =============================================================================

datagen = ImageDataGenerator(
    rotation_range=15,
    width_shift_range=0.10,
    height_shift_range=0.10,
    shear_range=0.10,
    zoom_range=0.10,
    brightness_range=(0.9, 1.1),
    fill_mode="nearest"
)

# Compute any internal statistics (safe even when no feature normalization is used)
datagen.fit(X_train)

print("=" * 60)
print("Data Augmentation Configuration")
print("=" * 60)

print(f"Rotation Range     : ±15°")
print(f"Width Shift Range  : ±10%")
print(f"Height Shift Range : ±10%")
print(f"Shear Range        : ±10%")
print(f"Zoom Range         : ±10%")
print(f"Brightness Range   : 0.9 - 1.1")
print(f"Fill Mode          : nearest")

print("\nData augmentation configured successfully.")
````

# Step 11: Build the EfficientNet Model

## Objective

Construct the convolutional neural network used for handwritten character recognition.

The model uses **EfficientNetB0** pretrained on ImageNet as the feature extractor, followed by a custom classification head for Odia characters.

---

## Architecture

```text
Input Image
      │
      ▼
EfficientNetB0
      │
      ▼
Global Average Pooling
      │
      ▼
Batch Normalization
      │
      ▼
Dropout
      │
      ▼
Dense Layer
      │
      ▼
Softmax Output
```

---

## Output

- Compiled TensorFlow model

---

## Notes

Transfer learning significantly reduces training time while improving model accuracy on relatively small datasets.

````python
# =============================================================================
# Build the EfficientNetB0 Model
# =============================================================================

# Load the pretrained EfficientNetB0 feature extractor
base_model = EfficientNetB0(
    weights="imagenet",
    include_top=False,
    input_shape=(IMAGE_SIZE, IMAGE_SIZE, 3),
    name="efficientnet_b0"
)

# Fine-tune all layers
base_model.trainable = True

# Build the classification model
model = Sequential(
    [
        base_model,

        GlobalAveragePooling2D(),

        BatchNormalization(),
        Dropout(0.50),

        Dense(
            256,
            activation="relu"
        ),

        BatchNormalization(),
        Dropout(0.50),

        Dense(
            NUM_CLASSES,
            activation="softmax"
        )
    ],
    name="lipy_ocr"
)

# Compile the model
model.compile(
    optimizer=Adam(
        learning_rate=LEARNING_RATE
    ),
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

print("=" * 60)
print("Model Configuration")
print("=" * 60)

print(f"Architecture      : EfficientNetB0")
print(f"Input Shape       : ({IMAGE_SIZE}, {IMAGE_SIZE}, 3)")
print(f"Output Classes    : {NUM_CLASSES}")
print(f"Learning Rate     : {LEARNING_RATE}")
print(f"Trainable Layers  : {'All'}")

print("\nModel Summary\n")

model.summary()
````

````python
# =============================================================================
# Configure Training Pipeline
# =============================================================================

# Local model directory
MODEL_ROOT.mkdir(parents=True, exist_ok=True)

checkpoint_path = MODEL_ROOT / "model.keras"
labels_path = MODEL_ROOT / "labels.json"
history_path = MODEL_ROOT / "training_history.json"

# Temporary checkpoint directory
TEMP_MODEL_DIR = MODEL_ROOT / ".training_tmp"
TEMP_MODEL_DIR.mkdir(parents=True, exist_ok=True)

temp_checkpoint_path = TEMP_MODEL_DIR / "best_model.keras"

# Optional Google Drive backup
DRIVE_MODEL_DIR = None

try:
    from google.colab import drive

    drive.mount("/content/drive", force_remount=False)

    DRIVE_MODEL_DIR = Path("/content/drive/MyDrive/lipy/models")
    DRIVE_MODEL_DIR.mkdir(parents=True, exist_ok=True)

except Exception:
    pass

# Load character mapping used by backend
try:
    from backend.labels import odia_ml_labels
except ImportError:
    odia_ml_labels = {}

label_to_char = {
    label: char
    for char, label in odia_ml_labels.items()
}

labels_data = [
    {
        "id": class_name,
        "char": label_to_char.get(class_name, "")
    }
    for class_name in valid_classes
]

labels_json = json.dumps(
    labels_data,
    ensure_ascii=False,
    indent=2
)

# Training callbacks
callbacks = [
    EarlyStopping(
        monitor="val_loss",
        patience=15,
        restore_best_weights=True
    ),

    ReduceLROnPlateau(
        monitor="val_loss",
        factor=0.5,
        patience=5,
        min_lr=1e-6
    ),

    ModelCheckpoint(
        filepath=str(temp_checkpoint_path),
        monitor="val_accuracy",
        mode="max",
        save_best_only=True
    )
]

print("=" * 60)
print("Training Pipeline Configured")
print("=" * 60)

print(f"Model Directory : {MODEL_ROOT}")
print(f"Checkpoint      : {checkpoint_path}")
print(f"Labels          : {labels_path}")
print(f"History         : {history_path}")
print(f"Callbacks       : {len(callbacks)}")
````

# Step 16: Test Random Prediction

## Objective

Perform inference on a randomly selected image from the test dataset.

This provides a qualitative evaluation of the trained model by comparing the predicted character with the ground truth label.

---

## Output

- Test image
- Predicted class
- Actual class
- Prediction confidence

---

## Notes

Predictions with confidence below the configured threshold are marked as **Uncertain**.

This step helps identify difficult samples and visually inspect model behavior.

````python
# =============================================================================
# Test Random Prediction
# =============================================================================

# Select a random sample
sample_index = random.randint(0, len(X_test) - 1)

sample_image = X_test[sample_index]

prediction = model.predict(
    np.expand_dims(sample_image, axis=0),
    verbose=0
)

predicted_index = np.argmax(prediction)
actual_index = np.argmax(y_test[sample_index])

confidence = float(np.max(prediction))

predicted_label = reverse_label_map[predicted_index]
actual_label = reverse_label_map[actual_index]

if confidence >= CONFIDENCE_THRESHOLD:
    prediction_text = f"{predicted_label} ({confidence:.2%})"
else:
    prediction_text = f"{predicted_label} (Uncertain: {confidence:.2%})"

plt.figure(figsize=(4, 4))
plt.imshow(sample_image)
plt.axis("off")

plt.title(
    f"Prediction : {prediction_text}\n"
    f"Actual     : {actual_label}"
)

plt.show()

print("=" * 60)
print("Random Prediction")
print("=" * 60)

print(f"Predicted Label : {predicted_label}")
print(f"Actual Label    : {actual_label}")
print(f"Confidence      : {confidence:.2%}")
````
